## 2. Teorik Temeller

---

### 2.1 Kombinatoriyel Optimizasyon Nedir?

Kombinatoriyel optimizasyon, ayrık (kesikli) ve sonlu karar değişkenlerine sahip problemlerde, bir amaç fonksiyonunu eniyilemek için yapılan hesaplamalı aramayı ifade eder. Bu tür problemler, çok sayıda kombinasyon içermeleri nedeniyle **NP-zor** sınıfına girer ve çözüm uzayı üstel olarak büyür.

Örnek problemler:

- Gezgin Satıcı Problemi (TSP)
- Vardiya planlaması
- Kaynak atama
- Sıralama ve çizelgeleme problemleri

Amaç genellikle aşağıdaki gibidir:

Minimize et:  
$ \min_{x \in \mathcal{X}} f(x) $  
Burada $ \mathcal{X} $ uygun çözümlerin (kısıtlar altında) kümesini, $ f(x) $ ise amaç fonksiyonunu ifade eder.

---

### 2.2 Matematiksel Temeller

Kombinatoriyel optimizasyonun temelleri, graf kuramı, doğrusal cebir ve olasılık kuramı üzerine kuruludur. Optimizasyon problemi, genel biçimiyle aşağıdaki gibi modellenebilir:

Amaç fonksiyonu:  
$ \min \; f(x) = \sum_{i} c_i x_i + \sum_{i < j} Q_{ij} x_i x_j $

Kısıtlar:  
$ x_i \in \{0, 1\} $

Burada:

- $ x_i $: ikili karar değişkenleri
- $ c_i $: doğrusal katsayılar
- $ Q_{ij} $: etkileşim katsayıları (ikili terimler)
- $ f(x) $: toplam maliyet veya skor

Bu yapı, QUBO formülasyonu ile doğrudan bağlantılıdır.

---

### 2.3 QUBO Formülasyonu

**QUBO (Quadratic Unconstrained Binary Optimization)**, ikili karar değişkenlerine sahip optimizasyon problemlerini çözmek için yaygın kullanılan bir formdur. QUBO problemi genel olarak şu şekilde tanımlanır:

Amaç:  
$ \min_{x \in \{0,1\}^n} x^T Q x $

Burada:

- $ x \in \{0,1\}^n $: ikili vektör (karar değişkenleri)
- $ Q \in \mathbb{R}^{n \times n} $: simetrik ağırlık matrisi
- $ x^T Q x $: toplam maliyet fonksiyonu

Bu yapı sayesinde birçok kombinatoriyel problem, kuantum çözücülerle çözülebilecek hale getirilir.

---

### 2.4 D-Wave's Ocean SDK Nedir?

Ocean SDK, D-Wave tarafından geliştirilen ve QUBO problemlerinin:

- Tanımlanmasını,
- Klasik çözücüler (Simulated Annealing, Tabu, vb.)
- Kuantum çözücüler (Quantum Annealing – QPU) ile çözülmesini

sağlayan açık kaynaklı bir yazılım geliştirme kitidir. Ocean SDK ile kullanıcılar, probleme özel `BinaryQuadraticModel` (BQM) oluşturabilir ve bunları `Sampler` sınıflarıyla çözebilir.

Başlıca bileşenler:

- `dimod`: QUBO modelleri ve çözüm sonuçları
- `neal`, `qbsolv`, `dwave.system`: klasik ve kuantum çözücüler
- `dwave.inspector`: çözüm görselleştirme aracı

Ocean SDK, akademik araştırmalardan endüstriyel uygulamalara kadar geniş bir kullanım alanına sahiptir.
